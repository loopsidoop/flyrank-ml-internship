## 1. My lane (or freestyle) and why

I will choose **Lane 2: Refresh / Content Opportunity Scoring**. The goal is to identify which pages represent the highest-value opportunities for a content refresh. I chose this because human review capacity is limited, and prioritizing the queue using observable signals rather than simple heuristics provides direct, measurable value. The starter baseline shows we can beat rules with a learned model.

```python
import pandas as pd
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')
print(f"Total pages in starter dataset: {len(df):,}")
```

## 2. The question: decision, action, cost of a wrong call

- **Decision:** Which pages should be reviewed first for refresh, expansion, or protection?
- **Unit of analysis:** The rows in the dataset.
- **Output:** A ranked queue of pages with the code explanining the decision.
- **Action:** A reviewer inspects the output (the ranked queue of pages) and decides whether to rewrite the title/meta, expand the content, or monitor it.
- **Cost of a wrong call:** Reviewer time is wasted on pages that have no real demand or are merely experiencing seasonal dips, while actual opportunities are ignored.

```python
# one row = one page.
unique_pages = df['content_id'].nunique()
print(f"Unique content IDs: {unique_pages:,} (matches total rows, confirming grain)")
```

## 3. Quick look at the data (2-3 real numbers)

Let's look at the dataset to see why this lane is worth pursuing. We'll check how many pages have strong traffic but are showing a "down" trend, representing prime opportunities for refresh.

```python
high_traffic = df[df['impressions_90d'] > 500]
declining = high_traffic[high_traffic['trend_direction'] == 'down']

print(f"Pages with >500 impressions in 90 days: {len(high_traffic):,}")
print(f"Of those, pages trending 'down': {len(declining):,} ({len(declining)/len(high_traffic):.1%})")
print("This shows a substantial subset of high-value pages that need review, perfectly matching a ranking problem.")
```

## 4. Careful words: what I can and can't claim

Our tool is designed to help human editors decide which pages to review by sorting them based on past traffic drops. It is an assistant, not a guarantee: we cannot prove that updating a page will definitely restore its traffic, nor are we trying to guess Google's secret ranking algorithm. We are simply using historical data to highlight which pages look like they need attention first.
